Treating Missing Values in Rows

```python
# As there are only 230 such records we can delete those.
# Note:- As we have total 307511 records deleting only 230 records will not affect our further analysis
applicationDF = applicationDF[(~(np.isnan(applicationDF['EXT_SOURCE_2']) & np.isnan(applicationDF['EXT_SOURCE_3'])))]
```

```python
# Comparing EXT_SOURCE_2 and EXT_SOURCE_3 column 
round(applicationDF.loc[:, ['EXT_SOURCE_2', 'EXT_SOURCE_3']].describe(),3)
```

We can see that variation from 25th to 75th percentile is equally distributed.

Also, the difference between mean and median is quite narrow for both the columns.

Thus, let's impute the missing values by the mean value of EXT_SOURCE_2 and EXT_SOURCE_3 respectively.

```python
# Imputing EXT_SOURCE_2 and EXT_SOURCE_3 by mean values
applicationDF.loc[np.isnan(applicationDF['EXT_SOURCE_2']), ['EXT_SOURCE_2']] = applicationDF['EXT_SOURCE_2'].mean()
applicationDF.loc[np.isnan(applicationDF['EXT_SOURCE_3']), ['EXT_SOURCE_3']] = applicationDF['EXT_SOURCE_3'].mean()
```

```python
(round(100*(applicationDF.isnull().sum()/len(applicationDF.index)),2))
```

```python
# Treating missing values in AMT_GOODS_PRICE
applicationDF['AMT_GOODS_PRICE'].describe().apply(lambda x: '%.2f' % x)
```

```python
applicationDF.boxplot(column =['AMT_GOODS_PRICE']) 
plt.show()
```

There are good amount of outliers in the column

Also We see the standard deviation of AMT_GOODS_PRICE is very high

So, imputing missing values with mean or median will be biased in this case

Hence, it is better to remove the rows with missing AMT_GOODS_PRICE. Infact these rows are only 0.09% of the whole records.

```python
applicationDF = applicationDF[~np.isnan(applicationDF['AMT_GOODS_PRICE'])]
```

We have 5 columns for the number of enquiries to Credit Bureau about the client. It doesn't make much sense to count the number for a very short period of time such as an hour or a day or a week before the application date. On the other had it also make lesser sense if we count the number for a very long period of time. Hence, it will be feasible for us if we take the count for a quater, which is not too short or too long time period. So, we have to do two things now:-

We will handle missing data for the column AMT_REQ_CREDIT_BUREAU_QRT

We will delete the column for the count of hour, day, week, month and year.

```python
# Handling AMT_REQ_CREDIT_BUREAU_QRT 
applicationDF['AMT_REQ_CREDIT_BUREAU_QRT'].describe()
```

This is an integer type variable, and thus have values 0, 1, 2 etc. 

We cannot impute the NaNs by the mean or the median (1.53 enquires does not make sense!). Thus, we need to impute them by the mode - the most common occurring value if that approach is feasible

```python
# converting AMT_REQ_CREDIT_BUREAU_QRT to type 'category'
applicationDF['AMT_REQ_CREDIT_BUREAU_QRT'] = applicationDF['AMT_REQ_CREDIT_BUREAU_QRT'].astype('category')

# displaying frequencies of each category
applicationDF['AMT_REQ_CREDIT_BUREAU_QRT'].value_counts()
```

The most common value of AMT_REQ_CREDIT_BUREAU_QRT is 0, so let's impute the NaNs by that.

```python
# Imputing AMT_REQ_CREDIT_BUREAU_QRT with 0
applicationDF.loc[pd.isnull(applicationDF['AMT_REQ_CREDIT_BUREAU_QRT']), ['AMT_REQ_CREDIT_BUREAU_QRT']] = 0
```

```python
# Deleting AMT_REQ_CREDIT_BUREAU_HOUR, AMT_REQ_CREDIT_BUREAU_DAY, AMT_REQ_CREDIT_BUREAU_WEEK, AMT_REQ_CREDIT_BUREAU_MON, AMT_REQ_CREDIT_BUREAU_YEAR
applicationDF =applicationDF.drop(['AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_YEAR'],axis=1)
```

```python
(round(100*(applicationDF.isnull().sum()/len(applicationDF.index)),2))
```

Removing outliers

```python
# Finding outliers in all the numerical columns with 1.5 IQR rule and removing the outlier records 
col_list_outlier = ['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE','AGE']
for col in col_list_outlier: 
    q1 = applicationDF[col].quantile(0.25)
    q3 = applicationDF[col].quantile(0.75)
    iqr = q3-q1
    range_low  = q1-1.5*iqr
    range_high = q3+1.5*iqr
    applicationDF = applicationDF.loc[(applicationDF[col] > range_low) & (applicationDF[col] < range_high)]

applicationDF.shape
```

```python
# Percentage of rows lost
round((100 - 100*(len(applicationDF.index)/307511)),2)
```

Though we lost around 10% rows in the outliers handling exercise, we got rid of the outliers for fair analysis of the data.

```python
applicationDF['AGE'].describe()
```

```python
# Define function for categorizing AGE_GROUP (Young, Mid age and Senior)
def age_group(x):
    if (x < 40):
        return 'Young'
    elif (x >= 40 and x < 60):
        return 'Mid Age'
    else:
        return 'Senior'
```

```python
#Creating new column AGE_GROUP
applicationDF['AGE_GROUP'] = applicationDF['AGE'].apply(age_group)
applicationDF.head()
```

Creating three credit amount groups - High, medium and low

```python
round(applicationDF['AMT_CREDIT'].describe(),2)
```

```python
# Define function for categorizing CREDIT_GROUP (High, Medium and Low)
def credit_group(x):
    if (x < 500000):
        return 'Low'
    elif (x >= 500000 and x < 750000):
        return 'Medium'
    else:
        return 'High'
```

```python
#Creating new column AGE_GROUP
applicationDF['CREDIT_GROUP'] = applicationDF['AMT_CREDIT'].apply(credit_group)
applicationDF.head()
```

Creating three income groups - Hign, medium and Low

```python
applicationDF['AMT_INCOME_TOTAL'].describe()
```

```python
# Define function for categorizing AMT_INCOME_TOTAL (Low, Medium and High)
def income_group(x):
    if (x < 100000):
        return 'Low'
    elif(x >= 100000 and x < 150000):
        return 'Medium'
    else:
        return 'High'
```

```python
# Creating column INCOME_GROUP with AMT_INCOME_TOTAL (Low, Medium and High)
applicationDF['INCOME_GROUP'] = applicationDF['AMT_INCOME_TOTAL'].apply(income_group)
applicationDF.head()
```
